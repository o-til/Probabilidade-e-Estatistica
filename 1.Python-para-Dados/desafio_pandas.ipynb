4. De acordo com os dados, advogar ou ser representante de vendas faz
você dormir menos? (Use o método ‘isin’, considere a média)

```python
attorney_and_sellers = df[df['Profissão'].isin(['Advogado(a)', 'Representante de vendas'])]
mean_sleep = attorney_and_sellers['Duração do sono'].mean()
mean_all = df['Duração do sono'].mean()
if mean_sleep < mean_all:
  print ("Ser advogado ou representante de vendas faz você dormir menos")
else:
  print ("Ser advogado ou representante de vendas não faz você dormir menos")
```

5. Entre quem fez enfermagem e quem fez medicina, quem tem menos
horas de sono? (Use o método ‘isin’, considere a média)

```python
doctors_and_nurses = df[df['Profissão'].isin(['Médico(a)', 'Enfermeiro(a)'])]
mean_doctors_and_nurses = doctors_and_nurses.groupby('Profissão')['Duração do sono'].mean()
if mean_doctors_and_nurses['Enfermeiro(a)'] < mean_doctors_and_nurses['Médico(a)']:
  print ("Quem fez enfermagem tem menos horas de sono, em média")
elif mean_doctors_and_nurses['Médico(a)']< mean_doctors_and_nurses['Enfermeiro(a)']:
  print ('Quem fez medicina tem menos horas de sono')
else:
  print ('As médias de horas são iguais, seja para medicina quando enfermagem')
```

7. Descubra qual a profissão menos frequente no conjunto. (Use
value_counts)

8. Quem tem maior pressão sanguínea média, homens ou mulheres?
(Considere a média)

```python
# Dividindo a coluna 'Pressão sanguínea' em duas colunas separadas
df[['Pressão sistólica', 'Pressão diastólica']] = df['Pressão sanguínea'].str.split('/', expand=True)

# Convertendo as novas colunas para números
df['Pressão sistólica'] = df['Pressão sistólica'].astype(int)
df['Pressão diastólica'] = df['Pressão diastólica'].astype(int)

print(df[['Pressão sistólica', 'Pressão diastólica']].head())

mean_blood_pressure = df.groupby('Gênero')['Pressão sistólica'].mean()

# Encontrar o gênero com maior média de pressão sanguínea
gender_pressure = mean_blood_pressure.idxmax()
max_pressure = mean_blood_pressure.max()

# Exibir o resultado
print(f"O gênero {gender_pressure} possui maior pressão sanguínea ({max_pressure:.2f})")
```

9. É predominante entre os participantes dormir 8 horas por dia
(considere usar Moda como medida)?

```python
mode_sleep = df['Duração do sono'].mode()
mode_total = mode_sleep.tolist()
if 8 in mode_sleep.values:
  print(f"A moda da duração de sono é: {mode_total}, logo, dormir 8 horas por dia é a moda entre os participantes")
else:
  print(f"A moda da duração de sono é: {mode_total}, logo, dormir 8 horas por dia NÃO é a moda entre os participantes")
```

10. Pessoas com frequências cardíacas acima de 70 dão mais passos que
pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
# Filtrando pessoas com frequência cardíaca acima de 70 e menor ou igual a 70
above_70 = df[df['Frequência cardíaca'] > 70]
below_or_equal_70 = df[df['Frequência cardíaca'] <= 70]

# Calculando a média de passos para cada grupo
mean_steps_above_70 = above_70['Passos diários'].mean()
mean_steps_below_or_equal_70 = below_or_equal_70['Passos diários'].mean()

# Exibindo os resultados
print(f"Média de passos para frequência cardíaca acima de 70: {mean_steps_above_70:.2f}")
print(f"Média de passos para frequência cardíaca menor ou igual a 70: {mean_steps_below_or_equal_70:.2f}")

# Comparando os resultados
if mean_steps_above_70 > mean_steps_below_or_equal_70:
    print("Pessoas com frequência cardíaca acima de 70 dão mais passos.")
elif mean_steps_above_70 < mean_steps_below_or_equal_70:
    print("Pessoas com frequência cardíaca menor ou igual a 70 dão mais passos.")
else:
    print("Ambos os grupos dão a mesma quantidade média de passos.")
```
